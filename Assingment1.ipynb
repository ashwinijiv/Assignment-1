```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [0,2,4,6,8]

## Resize the graph
## Plot a line graph
## Add title, labels, legend to your graph
## Save the Graph
plt.figure(figsize=(5,2),dpi=150)
plt.plot(x,y, label='kx', color='orange', linestyle='-.', marker='.',markersize=5, markeredgecolor='b',linewidth=1)

plt.xticks([1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])

plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.savefig('First_Graph.png', dpi=150)
plt.show()
```
